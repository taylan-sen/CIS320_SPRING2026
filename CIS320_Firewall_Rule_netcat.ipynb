### nc (netcat)

The **nc** shell command will allow a computer to receive or send packets over TCP or UDP. Effectively it allows the computer to act as a simple server or client.


    nc -l 12000             # listen on port 12000 using TCP (i.e. be a server)
    nc 35.173.12.31 12000   # connect over TCP to IP port (i.e. be a client

When using UDP instead of TCP, you need to add a -u option to both of the above commands.

    nc -l -u 12000             # listen on port 12000 using UDP (i.e. be a server)
    nc -u 35.173.12.31 12000   # connect over UDP to IP port (i.e. be a client
