# A Python Application for sending and receiving UDP and TCP packets

We will use a client/server architecture.

The 4 layers of the TCP/IP model are:
* **Application**
  * we will write a python application as the **client**, on Google Cloud, the professor will run the **server**
* **Transport**
  * there are two main protocols on this layer, we will use each, one at a time:
  * UDP - lightweight, fast, no connection
  * TCP - has connection, is reliable, slower
* **Internetworking**
* **Link**

<hr>

The python ***import*** keyword is used to pull in more functionality to the python environment. we will pull in networking functionality with the **socket** *module*.



```python
# CLIENT PROGRAM, WILL SEND A PACKET

import socket   # "pull in" networking functionality

print('UDP send application')

# we will need to provide two "addresses" in order to send a packet
# port and IP address


# CREATE (OPEN) THE SOCKET
#network_tcp_socket = socket.socket() # this creates a TCP socket object
network_udp_socket = socket.socket(type=socket.SOCK_DGRAM) # create UDP socket

# PREPARE MESSAGE
msg_str = 'Ty was here!'
msg_bytes = msg_str.encode()

# SEND PACKET
ip = '35.192.97.212'
port = 12000
network_udp_socket.sendto( msg_bytes, (ip, port)  )

network_udp_socket.close()
#network_tcp_socket.close()
```
