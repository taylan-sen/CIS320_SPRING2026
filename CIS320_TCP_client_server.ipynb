We implment the client side code for a TCP client server interaction

```python
import socket

print('starting tcpclient')
SERVER_IP = '34.58.243.152'
PORT = 12000

# OPEN SOCKET
s = socket.socket()

# CONNECT
s.connect(   (SERVER_IP, PORT)    )
# RECEIVE
msg = s.recv(2048)
print(msg.decode())
# SEND
# CLOSE
s.close()
```
